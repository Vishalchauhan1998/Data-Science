```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
train=pd.read_csv('hour.csv')
```

```python
ax = train.groupby('hr')['cnt'].sum().plot(kind='bar', figsize=(8, 6),
                                       legend = False, title ="Total Numbers of Bike Rentals counts by Hour", 
                                       color='blue', fontsize=12)
ax.set_xlabel("Hour", fontsize=12)
ax.set_ylabel("Counts", fontsize=12)
plt.show()
print(train.groupby('hr')['cnt'].sum())
```

```python
fig = plt.figure()
fig.suptitle('Boxplot for Humidity', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)
ax.boxplot(train['hum'])
ax.set_ylabel('Humidity')

q1=train['hum'].quantile(.25)
q3=train['hum'].quantile(.75)
iqr=q3-q1
print(iqr)
lowerlimit=q1-(1.5*iqr)
print(lowerlimit)
upperlimit=q3+(1.5*iqr)
print(upperlimit)
cnt=0
for i in train['hum']:
    if (i>upperlimit or i<lowerlimit):
        cnt=cnt+1
        print(i)
print("Count:",cnt)
```
